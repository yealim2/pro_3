```python
# 필요한 패키지들을 다 불러옵니다 
import numpy as np 
import pandas as pd 
import seaborn as sns
from matplotlib import pyplot as plt      
```

```python
# 데이터 이름은 vac으로 해주고 파일 불러오기 
vac= pd.read_csv('country_vaccinations.csv', parse_dates = ['date'])
```

```python
# 이 분석에서 필요 없는 열 을 지워줍니다. ('vaccines','source_name','source_website' 필요없음)
vac_1=vac.drop(['vaccines','source_name','source_website'],axis=1)
vac_1
```

위 vac의 실행 결과와 비교를 보면 'vaccines','source_name','source_website' 지워진거 확인 가능 

```python
vac_1.country.nunique() # 이 데이터 안에 223개국 있음 
```

백신접종률 하위순으로 나열, 예상대로 아프리카에 속한 국가 가장 낮았음 

```python
vac_1.nsmallest(100,'total_vaccinations')
```

상위 순으로 나열 역시 중국 

```python
vac_1.nlargest(100,'total_vaccinations')
```

```python
#전 국민 80% 이상 백신접종을 빨리 완료한 국가순으로  정렬 

vac_1_1st = vac.loc[vac['people_fully_vaccinated_per_hundred']>80,['country','date','people_fully_vaccinated_per_hundred']]
vac_1_1st.groupby(['country']).min().sort_values(by = 'date')
```

```python
#전 국민 대비 백신접종률이 20% 안되는 나라 별로 정렬 오름차순으로 
#c_under20 뜻은 country_vaccinated_under20% 라는 뜻 임 

v2 = vac_1.groupby(['country'])[['date', 'people_fully_vaccinated_per_hundred']].max()
c_under20 = v2.loc[v2['people_fully_vaccinated_per_hundred']<20,].sort_values(by = 'people_fully_vaccinated_per_hundred', ascending = True)
print(c_under20)
```

```python
#전 국민 대비 80퍼센트 이상 백신접종완료한 나라 정렬 내림차순으로 

v1 = vac_1.groupby(['country'])[['date', 'people_fully_vaccinated_per_hundred']].max()
v1.loc[v1['people_fully_vaccinated_per_hundred']>80,].sort_values(by = 'people_fully_vaccinated_per_hundred', ascending = False)
```

```python
#나라별 백신접종 현황을 그래프로 표현 
#Show vaccination rates countries

df_plot = vac_1.loc[:,['country', 'date', 'people_fully_vaccinated_per_hundred']]
grpA = ['South Korea','China', 'Afghanistan','United States', 'Germany', 'Japan', 'United Kingdom']
df_A = df_plot[df_plot['country'].isin(grpA)]

from matplotlib.dates import DateFormatter
import matplotlib.dates as mdates
fig, axes = plt.subplots(figsize = (20,8))
sns.lineplot(x = 'date', y = 'people_fully_vaccinated_per_hundred', hue = 'country', data = df_A, ax = axes, linewidth = 3) # linewidth 는 선 굵기 
axes.xaxis.set_major_formatter(DateFormatter("%y-%b")) # %b는 달, %y는 년도 
axes.set_xlabel("")
axes.set_ylabel("% of Vaccinated")
axes.set_title("vaccinatied of Contry")
axes.legend(title ="", prop = {'size':16})
```
