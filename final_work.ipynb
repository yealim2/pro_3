```python
import pandas as pd
import numpy as np
import zipfile
url = 'vaccination.zip'
```

zip파일 안에 있는 파일을 불러오기 위해 ZipFile 을 사용합니다 

```python
with zipfile.ZipFile(url) as z:
    print(z.namelist())
    kag = pd.read_csv(z.open('vaccinations.csv'))
    df = kag.iloc[1:]
```

'vaccinations.csv' 불러오기가 된것을 확인할 수 있습니다
